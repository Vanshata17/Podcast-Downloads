```python
import requests
from bs4 import BeautifulSoup
import os 
#for creating a file
import re
```

```python
rss_feed_url = 'https://lexfridman.com/feed/podcast/'
page = requests.get(rss_feed_url)
#print(page)
soup = BeautifulSoup(page.content, 'xml')
#soup
podcast_items=soup.find_all('item')
#podcast_items.find('description').text
```

```python
count = 0
output = []

for podcast in podcast_items:
    if count == 5:
        break
    
    title = podcast.find('title').text
    description = podcast.find('description').text  # Fetch the description for each podcast item
    mp3_url = podcast.find('enclosure')['url']
    sanitized_title = re.sub(r'[^\w\s.-]', '', title)
    
    if re.search(r'robot', description, re.I):
        mp3_file = requests.get(mp3_url)
        with open(f'./podcast_downloads/{sanitized_title}.mp3', 'wb') as f:
            f.write(mp3_file.content)
        
        output.append(title)
        count += 1
```

```python
def read_file(filename, chunk_size=5242880):
    with open(filename, 'rb') as _file:
        while True:
            data = _file.read(chunk_size)
            if not data:
                break
            yield data
```

```python

filenames=os.listdir("./podcast_downloads")
for file in filenames:
    file_path = f'./podcast_downloads/{file}'
    print("Uploading",  file)
    headers = {'authorization':  os.environ['ASSEMBLY_AI_KEY']}
    response = requests.post('https://api.assemblyai.com/v2/upload',
                             headers=headers,
                             data=read_file(file_path))
    print(response.json())

# # # Initialize the client with your API key
# client = assemblyai.Client('api_key')

# # Submit the audio file for transcription
# with open('audio_file.wav', 'rb') as f:
#     transcript = client.transcribe(f)

# # Get the transcription ID
# transcript_id = transcript['id']

# # Retrieve the transcription result
# result = client.transcript(transcript_id)

# # Get the transcript text
# transcript_text = result['text']

# print(transcript_text)
```
